```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import os
import glob
cwd = os.getcwd()
path = f"{cwd}/Data2017/preprocessed"
all_files = glob.glob(path + "/*.csv")
df = pd.DataFrame()
cur_df : pd.DataFrame
for file in all_files:
    cur_df = pd.read_csv(file)
    df = df.append(cur_df, ignore_index=True)

print(df.head())
X1 = df.iloc[:,1]   # Input feature 1
X2 = df.iloc[:,2]  # Input feature 2
y = df.iloc[:,5]    # Output vector
X = np.column_stack((X1, X2))   # Combine features to 2D Matrix
```

```python
fig = plt.figure(figsize=(12, 16))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(X1, X2, y, c="b", label="training data", marker="x", s=10)
ax.legend()
ax.set_zlabel("y")
plt.xlabel("Day")
plt.ylabel("Time")
plt.show()
```
