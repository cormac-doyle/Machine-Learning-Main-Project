```python
import pandas as pd
import numpy as np
import math
df = pd.read_csv("example.csv")
df.head()
```

```python
days = {"monday":1, "tuesday":2, "wednesday":3, "thursday":4, "friday":5, "saturday":6, "sunday":7}
# SPECIFY THE LOCATION OF THE DATA HERE
location = "Example"
```

```python
# 0 indicates north
# 1 indicates south


day = 0
date_string = ""
data_point_south  = []
data_point_north = []
output_df = pd.DataFrame()
temp_df : pd.DataFrame()
column_names = ["day", "time", "location", "direction", "volume"]
for index, row in df.iterrows():

    # Check if 'Total' exists in time column of current row,
    # if it does skip this iteration.
    if row["Time"] == "Total":
        continue

    # Check if the current row has a day specified
    if not pd.isna(row["Date"]):
        date_string = row["Date"]
        # Get the day as a string. Make it lowercase
        date_string = date_string.split()[0].lower()

    # Get the details to add to our new csv
    time = int(str(row["Time"]).replace(":", ""))
    data_point_north = [days[date_string], time, location, 0, row["N"]]
    data_point_south = [days[date_string], time, location, 1, row["S"]]
    temp_df = pd.DataFrame([data_point_north, data_point_south], columns=column_names)
    output_df = output_df.append(temp_df)

```
